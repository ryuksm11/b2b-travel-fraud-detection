```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("node_evaluation_table.csv")  # adjust path if needed
```

```python
df = df[df['fraud_flag'].notna()]
```

```python
df.shape
```

```python
df.describe()
```

```python
cols = ["anomaly_score_1", "anomaly_score_2", "anomaly_score_3"]

df["anomaly_score_avg"] = df[cols].mean(axis=1)
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('eval.csv')
```

```python
p95 = df["anomaly_score_avg"].quantile(0.90)

df_top95 = df[df["anomaly_score_avg"] >= p95]
```

```python
df_flag = df_top95[df_top95['fraud_flag'] == 1]

df_flag.shape
```
